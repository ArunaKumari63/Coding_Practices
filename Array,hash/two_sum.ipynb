<h1> Problem Statement </h1><small> Leet Code - Easy</small>

Given an array of integers <b> "nums" </b> and an integer <b> "target" </b>, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

<small> Example 1:

Input: nums = [2,7,11,15], target = 9

Output: [0,1]

Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].

Example 2:

Input: nums = [3,2,4], target = 6

Output: [1,2]

Example 3:

Input: nums = [3,3], target = 6

Output: [0,1]

Approach 1 - Brute Force

```python
def two_sum(nums,target):
    for i in range(len(nums)):
        for j in range(len(nums)):
            if i==j: # dont use same element twice
                j=j+1
                continue
            if target-nums[j]==nums[i]:
                return i,j
    return None
```

```python
nums = [2,7,11,15]
# nums = [10,8,4,2]
target = 9
output = two_sum(nums,target)
print(output)
```

Time Complexity - O(n^2)

Space Complexity - O(1)

Looping through each element in array to match if target-x=y and return index if there is a match, otherwise return None.

-----------------------------------------------------------------------

Approach 2 - Using One Pass Heap (to reduce time complexity)

```python
def two_sum_heap(nums,target):
    heap = {}
    for i in range(len(nums)):
        num_2 = target - nums[i]
        if num_2 in heap:
            return heap[num_2],i
        heap[nums[i]] = i
```

```python
nums = [4,-2,5,0,6,3,2,7]
target = 1
output = two_sum_heap(nums,target)
print(output)
```

Time Complexity - O(n)

Space Complexity - O(n)

Storing values in heap if target-x=y is not present in heap. If y exists in heap, return the index number
