<h1> Problem Statement </h1><small> Leet Code - Easy</small>

Write a function to find the longest common prefix string amongst an array of strings.

If there is no common prefix, return an empty string "".

<pre><small>
Example 1:

Input: strs = ["flower","flow","flight"]
Output: "fl"
Example 2:

Input: strs = ["dog","racecar","car"]
Output: ""
Explanation: There is no common prefix among the input strings.

```python
def longestCommonPrefix(strs) -> str:
        if len(strs)==0:
            return ""
        prefix = strs[0]
        for i in range(1,len(strs)):
            match_str=""
            if (len(strs[i])==len(prefix)) or (len(strs[i])<len(prefix)):
                itr=len(strs[i])
            elif len(strs[i])>len(prefix):
                itr=len(prefix)
            for j in range(itr):
                if prefix[j]!=strs[i][j]:
                    break
                else:
                    match_str = match_str+strs[i][j]
            prefix = match_str
        return prefix
```

```python
print(longestCommonPrefix(["cir","car"]))
```

Time Complexity: O(N*M) where N is number of strings and M is length of smallest string

Space Complexity: O(1)
